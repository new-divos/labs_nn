## Выбор ускорителя

**Google Colaboratory** - ресурс компании Google, который предоставляет бесплатные видеокарты для обучения нейронных сетей

Но важно в каждом ноутбуке **включить** эту опцию (задействовать видеокарту). В противном случае обучение нейронной сети будет происходить на процессоре, что займет значительно больше времени


Чтобы подключить ускоритель (видеокарту), необходимо выполнить следующее:
1. Среда выполнения → Сменить среду выполнения
2. В "Аппаратный ускоритель" выставить GPU или TPU
3. В "Форма среды выполнения" можно указать что требуется "Большой объём ОЗУ"
4. Текущий сеанс будет приостановлен и новая среда выполнения подключиться при запуске любой ячейки ноутбука


Для ускорения вычислений в Google Colaboratory можно подключать следующие ускорители:
* GPU (подключает графическую карту, хороша для ускорения обучения нейросетей)
* TPU (подключает графическую карту послабее, однако выделяет большой объём памяти - крайне полезно для работы с огромными обучающими базами, например при обработке текстов)
* None (не использует удалённую графическую карту)

При каждом запуске ноутбуке Colab'а, Вам будет выдаваться случайная видеокарта. На данный момент вы можете получить одну из следующий:
- Tesla K80 (минимальный вариант)
- Tesla P100 (средний вариант)
- Tesla V1000 (самая мощная видеокарта (если повезет:)

## Перезапуск среды выполнения

Если нужно "перезапустить" colab ноутбук - необходимо выполнить перезапуск среды выполнения (так как вычисления производятся на удалённом сервере, и даже отключение локальной машины сразу не завершит сеанс)

* Среда выполнения → Перезапустить среду выполнения
* ИЛИ сочетанием клавиш на английской раскладке: Ctrl + M + . (точка)

## Остановка среды выполнения

Если нужно завершить сеанс:

* Среда выполнения → Управление сеансами → кнопка ЗАВЕРШИТЬ напротив нужного сеанса
